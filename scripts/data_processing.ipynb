# standization of the columns


```python
 # 2. Standardize column names (snake_case is best practice for SQL later)
df.columns = df.columns.str.lower().str.replace(' ', '_').str.replace('-', '_')
```

# handling dates

```python
 # 3. Handle Dates
print("Parsing dates and handling missing values...")
# Using 'mixed' format as Kaggle CSVs sometimes have inconsistent date formats
df['order_date'] = pd.to_datetime(df['order_date'], format='mixed')
df['ship_date'] = pd.to_datetime(df['ship_date'], format='mixed')
```

```python
df.head()
```

```python
 # 4. Handle Nulls (Drop rows where critical fields like Sales or Dates are missing)
df = df.dropna(subset=['order_date', 'sales', 'category'])
```

```python
# 5. Localization: Redefining the Geography
print("Applying Mumbai localization mapping...")
    
    # Set static country and state
df['country'] = 'India'
df['state'] = 'Maharashtra'

df.head()
```

# perform mapping 

```python
# FIX: Randomly distribute our exact Mumbai zones across the dataset 
# to completely avoid any 'NaN' or missing region mapping issues.
mumbai_zones = ['Western Suburbs', 'Eastern Suburbs', 'South Mumbai', 'Thane & Navi Mumbai']
df['region'] = np.random.choice(mumbai_zones, size=len(df))

# Map local areas to those specific zones
mumbai_areas = {
    'Western Suburbs': ['Bandra', 'Andheri', 'Borivali', 'Goregaon', 'Malad'],
    'Eastern Suburbs': ['Powai', 'Ghatkopar', 'Kurla', 'Vikhroli', 'Mulund'],
    'South Mumbai': ['Colaba', 'Nariman Point', 'Lower Parel', 'Dadar', 'Worli'],
    'Thane & Navi Mumbai': ['Thane West', 'Vashi', 'Belapur', 'Kalyan', 'Airoli']
}

# Assign city safely based on the newly assigned guaranteed region
df['city'] = df['region'].apply(lambda r: np.random.choice(mumbai_areas[r]))

# Clean up irrelevant columns
if 'postal_code' in df.columns:
    df = df.drop(columns=['postal_code'])
```
